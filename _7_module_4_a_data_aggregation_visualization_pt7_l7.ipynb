```python
fund_df.groupby(['fund_owner_name', 'fund_type'])[['management_fee']].agg(['count', 'min', 'median', 'max'])
```

```python
# Đơn vị nav là tỷ vnđ 'nav':'sum'
fund_owner_stats = fund_df.groupby(['fund_owner_name'])\
    .agg({'fund_type':pd.Series.nunique, 'nav_change_12m':'median'})\
    .sort_values(by='nav_change_12m', ascending=False)

fund_owner_stats\
    .style.format("{:,.1f}")\
    .background_gradient(cmap='Spectral',
                         axis=1,
                         subset=['nav_change_12m'],
                         vmin=fund_owner_stats['nav_change_12m'].min(),
                         vmax=fund_owner_stats['nav_change_12m'].max())
```

## Pivot table

### Tóm tắt lý thuyết

```python
import pandas as pd
pd.pivot_table(data, index=None, values=None,  columns=None,
               aggfunc='mean', fill_value=None, margins=False,
               dropna=True, margins_name='All', observed=False)
```

| Tên chức năng  | Mô tả                                                                                           |
| :------------- | :---------------------------------------------------------------------------------------------- |
| `values`       | Tên cột hoặc danh sách tên các cột để tổng hợp; mặc định tổng hợp tất cả các cột số                        |
| `index`        | Tên cột hoặc danh sách tên các cột để nhóm theo các hàng của bảng pivot kết quả                     |
| `columns`      |Tên cột hoặc danh sách tên các cột để nhóm theo các cột của bảng pivot kết quả                      |
| `aggfunc`      | Hàm tổng hợp hoặc danh sách các hàm (`'mean'` mặc định); có thể là bất kỳ hàm nào hợp lệ trong ngữ cảnh groupby |
| `fill_value`   | Chỉ định giá trị sẽ thay thế các giá trị bị thiếu (NaN) trong bảng kết quả                                                 |
| `dropna`       | Nếu `True`, không bao gồm các cột mà tất cả các dòng đều là `NA`                             |
| `margins`      | Thêm tổng phụ và tổng chung cho hàng/cột (`False` mặc định)                                     |
| margins_name   | Đặt tên của hàng/cột tổng (mặc định = 'All')                                                        |
| observed       | Chỉ dành cho dữ liệu phân loại – nếu True sẽ chỉ hiển thị các giá trị quan sát được cho các nhóm phân loại (mặc định = False) |


Bảng dưới đây thể hiện danh sách các hàm tính toán cơ bản của Pandas, có thể sử dụng hàm tính toán bất kỳ của thư viện khác như Numpy

|  #   |   Chức năng    |               Mô tả                |
| :--: | :------------: | :--------------------------------: |
|  1   |     count      | Số lượng quan sát không rỗng (Null/NaN)       |

|  7   |      min       | Giá trị nhỏ nhất                   |
|  8   |      max       | Giá trị lớn nhất                   |
|  9   |      abs       | Giá trị tuyệt đối                  |
|  10  |      prod      | Tích của các giá trị               |
|  11  |     cumsum     | Tổng lũy kế                        |
|  12  |    cumprod     | Tích lũy kế                        |

### Minh họa

```python
pd.pivot_table(screener_df,
               index=['industryName.en'],
               values='ticker',
               columns='exchangeName.en',
               aggfunc=pd.Series.nunique,
               fill_value=0,
               margins=True,
               margins_name = 'Grand Total'
               )
```

```python
pd.pivot_table(screener_df,
               index=['exchangeName.en', 'industryName.en'],
               values='ticker',
               aggfunc=pd.Series.nunique,
               fill_value=0,
               margins=True,
               margins_name = 'Grand Total')
```

```python
# Trung vị % chia cổ tức của các công ty theo ngành và sàn

pd.pivot_table(screener_df,
               index=['exchangeName.en', 'industryName.en'],
               values='dividendYield',
               aggfunc='median',
               fill_value=0
               )
```

```python
# Trung vị % chia cổ tức của các công ty theo ngành và sàn
pd.pivot_table(screener_df,
               index='industryName.en',
               columns='exchangeName.en',
               values='dividendYield',
               aggfunc='median',
               fill_value=0
               )
```

```python
# Trung vị % chia cổ tức của các công ty theo ngành và sàn
pd.pivot_table(screener_df,
               index='industryName.en',
               columns='exchangeName.en',
               values='dividendYield',
               aggfunc=['min', 'median', 'max'],
               fill_value=0
               )
```

```python
# Trung vị % chia cổ tức của các công ty theo ngành và sàn
pd.pivot_table(screener_df,
               index=['exchangeName.en', 'industryName.en'],
               values=['dividendYield', 'ticker'],
               aggfunc={'ticker': pd.Series.nunique, 'dividendYield':'median'},
               fill_value=0,
               margins=True,
               margins_name='Tổng hợp'
               )
```

## Crosstab

```python
pd.crosstab(index=screener_df['exchangeName.en'],
            columns=screener_df['industryName.en'],
            values=screener_df['dividendYield'],
            aggfunc='median')
```

### a. Visualize Amounts | Biểu diễn số lượng

Cách tiếp cận phổ biến nhất khi biểu diễn số lượng (vd: giá trị số biểu thị cho 1 số categrory) là sử dụng bar chart (biểu đồ cột) dù là theo trục ngang hay dọc. Tuy nhiên thay vì sử dụng cột, chúng ta cũng có thể biểu diễn các chấm (dot) ở các vị trí tương ứng đỉnh của cột.

Nếu có 2 hoặc nhiều category cần phải biểu diễn số lượng, chúng ta có thể nhóm hoặc xếp chồng các cột với nhau. Chúng ta cũng có thể sử dụng dạng heatmap sử dụng trục biểu diễn số theo cả 2 chiều x và y.

### b. Visualize Distributions | Biểu diễn phân phối dữ liệu

Histogram (phân phối) và Density plot (mật độ) cho phép biểu diễn phân phối (distribution) trực quan nhất tuy nhiên cả 2 loại biểu đồ này yêu cầu lựa chọn tham số và đôi khi dễ gây hiểu nhầm. Cumulative densities và quantile-quantile (q-q) plots luôn thể hiện dữ liệu một cách trung thực nhưng đôi khi khó giải thích hơn.

![](https://thinhvu.com/wp-content/uploads/python_course/viz_distribution_1.png)
Boxplot, violin plot, strip plot và sina plot hữu ích khi chúng ta muốn biểu diễn nhiều phân phối cùng một lúc và / hoặc nếu chúng ta chủ yếu quan tâm đến sự thay đổi tổng thể giữa các phân bố. Stacked histogram và overlapping density cho phép so sánh sâu hơn với số lượng phân phối nhỏ hơn, mặc dù stacked histogram có thể khó giải thích và tốt nhất nên tránh. Biểu đồ Ridgeline có thể là một sự thay thế hữu ích cho các violin plot và thường hữu ích khi biểu diễn số lượng rất lớn các phân bố hoặc những thay đổi trong phân bố theo thời gian.
![](https://thinhvu.com/wp-content/uploads/python_course/viz_distribution_2.png)

### c. Visualize Proportions | Biểu diễn tỉ lệ

Số liệu dạng tỷ lệ có thể được biểu diễn bằng Pie chart, Bars hoặc Stacked Bars. Đối với số lượng, khi chúng ta biểu diễn tỷ lệ bằng Bars chart, các thanh có thể được sắp xếp theo chiều dọc hoặc chiều ngang. Pie chart nhấn mạnh rằng các phần riêng lẻ cộng lại thành tổng thể và làm nổi bật các phân số đơn giản. Tuy nhiên, các phần riêng lẻ dễ dàng so sánh hơn trong các thanh cạnh nhau. Các thanh xếp chồng lên nhau trông khó hiểu đối với một tập hợp tỷ lệ, nhưng có thể hữu ích khi so sánh nhiều tập hợp tỷ lệ.
![](https://thinhvu.com/wp-content/uploads/python_course/viz_proportion_1.png)
Khi biểu diễn nhiều tập hợp tỷ lệ hoặc thay đổi tỷ lệ trong các điều kiện, Pie chart có xu hướng không hiệu quả về không gian và thường che khuất các mối quan hệ. Group Bars hoạt động tốt miễn là số lượng điều kiện được so sánh là vừa phải và Stacked Bars có thể hoạt động với số lượng lớn các điều kiện. Stacked Densities thích hợp khi tỷ lệ thay đổi theo một biến số liên tục.
![](https://thinhvu.com/wp-content/uploads/python_course/viz_proportion_2.png)
Khi tỷ lệ được chỉ định theo nhiều biến nhóm, Mosaic Plot, Treemap hoặc Parallel Sét là các phương pháp tiếp cận trực quan hữu ích. Mosaic Plot giả định rằng mọi cấp của một biến nhóm có thể được kết hợp với mọi cấp của biến nhóm khác, trong khi biểu đồ dạng treemap không đưa ra giả định như vậy. Treemap hoạt động tốt ngay cả khi các phần của một nhóm hoàn toàn khác biệt với các phần của nhóm khác. Parallel Sét hoạt động tốt hơn so với Mosaic Plot hoặc Treemap khi có nhiều hơn hai biến nhóm.
![](https://thinhvu.com/wp-content/uploads/python_course/viz_proportion_3.png)

### d. Scatterplots | Biểu diễn tương quan

Scatterplots thể hiện biểu diễn của nguyên mẫu khi chúng ta muốn hiển thị một biến định lượng so với một biến định lượng khác. Nếu chúng ta có ba biến định lượng, chúng ta có thể map một biến vào kích thước chấm, tạo ra một biến thể của Scatterplots được gọi là Bubble Chart. Đối với dữ liệu dạng cặp, trong đó các biến dọc theo trục x và y được đo bằng các đơn vị giống nhau, nói chung sẽ hữu ích khi thêm một dòng biểu thị x = y. Dữ liệu dạng cặp cũng có thể được hiển thị dưới dạng đồ thị độ dốc của các điểm được ghép nối với nhau bằng các đường thẳng.

![](https://thinhvu.com/wp-content/uploads/python_course/viz_relationship_1.png)
Đối với số lượng lớn các điểm, Scatterplot thông thường có thể trở nên không hữu ích do biểu đồ có quá dày đặc (không phân biệt được các chấm nữa). Trong trường hợp này, Density Contours, 2D Bins hoặc Hex Bins có thể cung cấp một giải pháp thay thế. Mặt khác, khi chúng ta muốn hình dung nhiều hơn hai đại lượng, chúng ta có thể chọn vẽ biểu đồ các hệ số tương quan dưới dạng một biểu đồ tương quan thay vì dữ liệu thô cơ bản.
![](https://thinhvu.com/wp-content/uploads/python_course/viz_relationship_2.png)
Khi trục x đại diện cho thời gian hoặc một đại lượng gia tăng liên tục chẳng hạn như liều điều trị, chúng ta thường vẽ Line graph. Nếu chúng ta có một chuỗi thời gian của hai biến phản hồi (response variables), chúng ta có thể vẽ Connected Scatterplot, khi đó trước tiên biểu đồ của hai biến phản hồi được vẽ trong Scatterplot và sau đó kết nối các dấu chấm tương ứng với các điểm thời gian liền kề. Chúng ta có thể sử dụng Smooth Line Graph để biểu diễn các xu hướng trong một tập dữ liệu lớn hơn.
![](https://thinhvu.com/wp-content/uploads/python_course/viz_relationship_3.png)

```python
# @title ECharts Candlestick
from pyecharts import options as opts
from pyecharts.charts import Kline, Bar, Line

COLOR_BULL = "#26a69a"
COLOR_BEAR = "#ef5350"


def candlestick_chart(data, title='Giá lịch sử của cổ phiếu', width=1700, height=800):
    """
    Biểu diễn dữ liệu đồ thị nến với pyecharts trong môi trường Google Colab

    Tham số:
        - data: DataFrame dữ liệu OHLCV tiêu chuẩn, tiêu đề định dạng lower case bao gồm các cột: open, high, low, close, volume
        - title: Tiêu đề của đồ thị
    """
    candle_data = candle_df[['Open', 'Close', 'Low', 'High']].values.tolist()
    volume_data = candle_df[['Volume', 'Label']].values.tolist()
    time_series = candle_df.index.strftime('%Y-%m-%d').to_list()

    kline = (
        Kline(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
        .add_xaxis(xaxis_data=time_series)
        .add_yaxis(
            series_name="OHLC",
            y_axis=candle_data,
            itemstyle_opts=opts.ItemStyleOpts(color=COLOR_BEAR, color0=COLOR_BULL),
        )
        .set_global_opts(
            legend_opts=opts.LegendOpts(
                is_show=False, pos_bottom=10, pos_left="center"
            ),
            datazoom_opts=[
                opts.DataZoomOpts(
                    is_show=False,
                    type_="inside",
                    xaxis_index=[0, 1],
                    range_start=80,
                    range_end=100,
                ),
                # opts.DataZoomOpts(
                #     is_show=True,
                #     xaxis_index=[0, 1],
                #     type_="slider",
                #     pos_top="90%",
                #     range_start=80,
                #     range_end=100,
                # ),
            ],
            yaxis_opts=opts.AxisOpts(
                is_scale=True,
                splitarea_opts=opts.SplitAreaOpts(
                    is_show=True, areastyle_opts=opts.AreaStyleOpts(opacity=1)
                ),
                # name_location="end",
            ),
            xaxis_opts=opts.AxisOpts(
                axislabel_opts=opts.LabelOpts(is_show=True), # Show date on x-axis
            ),

            tooltip_opts=opts.TooltipOpts(
                trigger="axis",
                axis_pointer_type="cross",
                background_color="rgba(245, 245, 245, 0.8)",
                border_width=1,
                border_color="#ccc",
                textstyle_opts=opts.TextStyleOpts(color="#000"),
            ),
            visualmap_opts=opts.VisualMapOpts(
                is_show=False,
                dimension=2,
                series_index=5,
                is_piecewise=True,
                pieces=[
                    {"value": 1, "color": COLOR_BULL},
                    {"value": -1, "color": COLOR_BEAR},
                ],
            ),
            axispointer_opts=opts.AxisPointerOpts(
                is_show=True,
                link=[{"xAxisIndex": "all"}],
                label=opts.LabelOpts(background_color="#777"),
            ),
            brush_opts=opts.BrushOpts(
                x_axis_index="all",
                brush_link="all",
                out_of_brush={"colorAlpha": 0.1},
                brush_type="lineX",
            ),
        )
    )

    bar = (
            Bar()
            .add_xaxis(xaxis_data=time_series)
            .add_yaxis(
                series_name="Volume",
                y_axis=[{
                    "value": vol[0],
                    "itemStyle": {"color": COLOR_BULL if vol[1] > 0 else COLOR_BEAR} # Hiện màu BULL nếu cột thứ 2 trong list dữ liệu volume (gồm 2 cột, cột 1 là giá trị, cột 2 set 1 hoặc -1)
                } for vol in volume_data],
                xaxis_index=1,
                yaxis_index=1,
                label_opts=opts.LabelOpts(is_show=False),
            )
            .set_global_opts(
                xaxis_opts=opts.AxisOpts(
                    type_="category",
                    is_scale=True,
                    grid_index=1,
                    boundary_gap=False,
                    axisline_opts=opts.AxisLineOpts(is_on_zero=False),
                    axistick_opts=opts.AxisTickOpts(is_show=False),
                    splitline_opts=opts.SplitLineOpts(is_show=False),
                    axislabel_opts=opts.LabelOpts(is_show=False), # Show date on x-axis
                    split_number=20,
                    min_="dataMin",
                    max_="dataMax",
                ),
                yaxis_opts=opts.AxisOpts(
                    grid_index=1,
                    is_scale=True,
                    split_number=2,
                    axislabel_opts=opts.LabelOpts(is_show=True),
                    axisline_opts=opts.AxisLineOpts(is_show=False),
                    axistick_opts=opts.AxisTickOpts(is_show=False),
                    splitline_opts=opts.SplitLineOpts(is_show=False),
                ),
                legend_opts=opts.LegendOpts(is_show=False),
        )
        )


    # Grid Overlap + Bar
    grid_chart = Grid(
        init_opts=opts.InitOpts(
            width="1500px",
            height="800px",
            animation_opts=opts.AnimationOpts(animation=True),
        )
    )


    grid_chart.add(
        kline,
        grid_opts=opts.GridOpts(pos_left="10%", pos_right="8%", height="65%"),
    )


    grid_chart.add(
        bar,
        grid_opts=opts.GridOpts(
            pos_left="10%", pos_right="8%", pos_top="63%", height="16%"
        ),
    )
# c.render_notebook() # Sử dụng với Jupyter Notebook
    return pn.pane.ECharts(grid_chart, width=width, height=height) # Dùng kèm với Panel cho Colab hoặc Notebook bất kỳ
```
