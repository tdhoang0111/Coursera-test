# II. Giới thiệu HTML

> HTML (HyperText Markup Language) là ngôn ngữ đánh dấu dùng để tạo cấu trúc cho các trang web. Một trang HTML bao gồm nhiều thẻ (tags) để mô tả các phần tử như tiêu đề, đoạn văn, liên kết, hình ảnh,... Đây là tiêu chuẩn tạo hiển thị nội dung trên nền tảng internet của các website hiện nay.

## 2.1. Cấu trúc của trang web HTML

````python
<!DOCTYPE html>
<html>
<head>
    <title>Tiêu đề trang</title>
</head>
<body>
    <h1>Tiêu đề chính</h1>
    <p>Đây là một đoạn văn bản.</p>
    <a href="https://example.com">Đây là một liên kết</a>
    <img src="image.jpg" alt="Mô tả hình ảnh">
</body>
</html>
````

Nói một cách dễ hiểu, bạn có thể hình dung các thẻ (tag) trong HTML sử dụng để đánh dấu đâu là các phần nội dung cụ thể trên 1 website. Các thẻ thường đi theo cặp. Ví dụ <head></head> cho phần thiết lập đầu trang, <body></body> cho nội dung chính của trang, <a></a> cho phần liên kết (URL), <p></p> cho văn bản, vv
Hiểu được cấu trúc của HTML và cách thức 1 website hoạt động sẽ cho phép bạn chọn đúng đối tượng chứa thông tin mình cần trích xuất.

# III. Giới thiệu CSS

> CSS (Cascading Style Sheets) là ngôn ngữ dùng để mô tả giao diện và định dạng của một tài liệu viết bằng HTML. CSS giúp tách biệt nội dung và hình thức của trang web, cho phép bạn kiểm soát cách các phần tử HTML hiển thị trên trang.

**Mối quan hệ của HTML & CSS**

1. **HTML** được xem như là khung xương hoặc cấu trúc của một ngôi nhà. Khi bạn xây dựng một ngôi nhà, bạn cần có các yếu tố cơ bản như tường, mái, cửa, và cửa sổ. HTML đóng vai trò tương tự bằng cách cung cấp cấu trúc cơ bản cho một trang web, bao gồm các thẻ như `<div>`, `<p>`, `<h1>`, `<a>`, `<img>`, v.v.

2. **CSS** giống như việc trang trí, sơn sửa và làm đẹp ngôi nhà. Sau khi bạn đã xây dựng khung xương của ngôi nhà bằng HTML, bạn sẽ cần phải sơn tường, đặt đồ nội thất, và thêm các chi tiết trang trí để ngôi nhà trở nên đẹp mắt và hấp dẫn. CSS giúp bạn làm điều này cho trang web bằng cách kiểm soát giao diện và bố cục của các phần tử HTML.

## 3.1. Cấu trúc của CSS

CSS có thể được khai báo trong HTML bằng ba cách: inline (nội tuyến), internal (nội bộ), và external (bên ngoài).

- **Inline:** CSS được khai báo trực tiếp trong thẻ HTML sử dụng thuộc tính style.

```
<p style="color: blue; font-size: 20px;">Đoạn văn màu xanh dương.</p>
```

- **Internal**: CSS được khai báo trong phần `<head>` của tài liệu HTML sử dụng thẻ `<style>`.

```
<!DOCTYPE html>
<html>
<head>
    <style>
        p {
            color: blue;
            font-size: 20px;
        }
    </style>
</head>
<body>
    <p>Đoạn văn màu xanh dương.</p>
</body>
</html>
```

- **External**: CSS được khai báo trong một file riêng biệt và liên kết với tài liệu HTML sử dụng thẻ <link>.

```
<!DOCTYPE html>
<html>
<head>
    <link rel="stylesheet" type="text/css" href="styles.css">
</head>
<body>
    <p>Đoạn văn màu xanh dương.</p>
</body>
</html>
```

file `style.css` sẽ có nội dung dạng này:

```
p {
    color: blue;
    font-size: 20px;
}
```

## 3.2. Chọn phần tử CSS

Tương tự với thẻ HTML thì đoạn mã CSS cũng sử dụng để tham chiếu và mô tả thuộc tính của một phần tử bất kỳ xuất hiện trên trang web. Do đó, sử dụng thuộc tính CSS cũng có thể chọn được phần tử mà bạn hướng đến và qua đó trích xuất thông tin bạn cần. Dưới đây là một số cách thường được sử dụng để chỉ định một phần tử CSS.

### a. CSS Selector (Bộ chọn CSS)

Bộ chọn CSS được sử dụng để chọn các phần tử HTML mà bạn muốn áp dụng các quy tắc CSS.

**1. Bộ chọn phần tử (Element Selector)**

Chọn tất cả các phần tử cùng loại.

```
p {
    color: blue;
}
```

2. **Bộ chọn lớp (Class Selector)**

Chọn các phần tử có thuộc tính class cụ thể.

Đoạn mã HTML dưới đây được phân nhóm thuộc tính sử dụng từ khóa `class`.
```
<p class="blue-text">Đoạn văn màu xanh dương.</p>
```

theo đó `blue-text` sẽ được định nghĩa trong phần CSS như sau:

```
.blue-text {
    color: blue;
}
```

**2. Bộ chọn ID (ID Selector)**

Chọn phần tử có thuộc tính id cụ thể.

Nội dung HTML
```
<p id="main-text">Đoạn văn chính.</p>
```

Nội dung CSS đi kèm

```
#main-text {
    color: blue;
}
```

**3. Bộ chọn thuộc tính (Attribute Selector)**

Chọn các phần tử dựa trên thuộc tính và giá trị thuộc tính.

```
<a href="https://example.com" target="_blank">Liên kết</a>
```

Nội dung CSS đi kèm
```
a[target="_blank"] {
    color: red;
}
```

**4. Bộ chọn nhóm (Group Selector)** - Nâng cao

Mã CSS quy định 1 kiểu hiển thị chung cho nhóm các phần tử:

```
h1, h2, p {
    color: blue;
}
```

**5. Bộ chọn con (Child Selector)**

Chọn các phần tử con trực tiếp của một phần tử cụ thể.

```
<div>
    <p>Đoạn văn bên trong div.</p>
</div>
```

Nội dung CSS
```
div > p {
    color: blue;
}
```

**6. Bộ chọn tổng quát (Universal Selector)**

Chọn tất cả các phần tử.

```
* {
    color: blue;
}
```

**7. Bộ chọn giả (Pseudo-selector)**

Chọn các trạng thái cụ thể của phần tử, như khi phần tử được hover, focus,...

```
a:hover {
    color: red;
}
```

#### Sử dụng CSS Selectors

````python
from bs4 import BeautifulSoup

html_doc = """
<!DOCTYPE html>
<html>
<head>
    <title>Tiêu đề trang</title>
</head>
<body>
    <h1 class="title">Tiêu đề chính</h1>
    <p class="content">Đoạn văn nội dung.</p>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Tìm phần tử bằng class
title = soup.select_one('.title').text
print("Tiêu đề:", title)

# Tìm phần tử bằng tag
content = soup.select_one('p').text
print("Nội dung:", content)
````

### b. Xpath

> XPath (XML Path Language) là ngôn ngữ để xác định các phần tử trong một tài liệu XML hoặc HTML. Nó thường được sử dụng trong các công cụ web scraping như Selenium và Scrapy.

#### Cú pháp cơ bản của XPath

```
            <p>Đoạn văn</p>
        </div>

</html>
```

chúng ta sử dụng cú pháp Xpath:

Bạn có thể thấy cú pháp Xpath tương tự như đường dẫn của file trong cây thư mục hệ thống. Sau mỗi dấu `/` là một cấu độ phần tử nhỏ hơn.

**2. Chọn phần tử theo đường dẫn tương đối**

````python
//div/p
````

**3. Chọn phần tử theo thuộc tính**

````python
<a href="https://example.com">Liên kết</a>
````

Sử dụng Xpath sau:

```
//a[@href="https://example.com"]
```

**4. Chọn phần tử theo text**

Nội dung HTML

```
<p>Đoạn văn nội dung.</p>
```

Xpath tương ứng:

```
//p[text()="Đoạn văn nội dung."]
```

#### Sử dụng Selenium với XPath

````python
from selenium import webdriver

driver = webdriver.Chrome()
driver.get("https://vnexpress.net/")

# Tìm phần tử bằng XPath
element = driver.find_element_by_xpath('//h1')
print("Tiêu đề:", element.text)

driver.quit()
````

#### Sử dụng Scrapy với XPath

````python
import scrapy

class ExampleSpider(scrapy.Spider):
    name = "example"
    start_urls = [
        'https://vnexpress.net/',
    ]

    def parse(self, response):
        # Tìm phần tử bằng XPath
        title = response.xpath('//h1/text()').get()
        print("Tiêu đề:", title)
````

## 3.3. Sử dụng HTML/CSS trong Web Scraping như thế nào?

Trong Web Scraping, bạn cần nắm rõ được phần tử chứa dữ liệu mình cần trích xuất trên giao diện web là đối tượng nào để chỉ định các thư viện python như BeautifulSoup, Scrapy, hay Selenium, Playwright trích xuất dữ liệu chính xác cho bạn.

Dưới đây là một ví dụ giúp bạn hình dung cách thức lựa chọn phần tử và truy xuất dữ liệu HTML với Python. Trong đó:
- `html_doc` là biến chứa dữ liệu HTML dạng văn bản, trong thực tế bạn sẽ sử dụng các thư viện để đọc file html vào biến (tương tự như `html_doc` đang giữ thông tin toàn bộ trang web cần truy xuất).
- `BeautifulSoup` là thư viện giúp phân giải (parse) thông tin HTML và trả về đối tượng văn bản/nội dung bạn cần tách ra từ trang web hoàn chỉnh.
- `title`, `p`, `a` dưới dạng thuộc tính của đối tượng `soup` chính là các thẻ chứa nội dung tiêu đề, văn bản và liên kết trong trang. Bạn có thể hình dung rằng chúng ta đang bóc tách thông tin theo lớp dưới dạng 1 cây thư mục trong đó website (hay đối tượng soup) là thư mục cha, cách truy cập title, p, hay a chính là các thư mục con chứa thông tin và bản thân thông tin chính là 1 file trong thư mục con đó.

````python
from bs4 import BeautifulSoup

html_doc = """
<!DOCTYPE html>
<html>
<head>
    <title>Tiêu đề trang</title>
</head>
<body>
    <h1>Tiêu đề chính</h1>
    <p>Đây là một đoạn văn bản.</p>
    <a href="https://example.com">Đây là một liên kết</a>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Lấy tiêu đề trang
title = soup.title.string
print("Tiêu đề trang:", title)

# Lấy nội dung đoạn văn bản
paragraph = soup.p.string
print("Đoạn văn:", paragraph)

# Lấy liên kết
link = soup.a['href']
print("Liên kết:", link)
````

````python
import requests
from bs4 import BeautifulSoup

# Địa chỉ URL của bài báo
# url = "https://vnexpress.net/gia-dat-con-len-den-dau-4786448.html"
url = "https://vnexpress.net/lam-viec-12-tieng-moi-ngay-4785933.html"

# Gửi yêu cầu GET đến URL
response = requests.get(url)

# Kiểm tra nếu yêu cầu thành công
if response.status_code == 200:
    # Tạo đối tượng BeautifulSoup để phân tích nội dung HTML
    soup = BeautifulSoup(response.text, 'html.parser')

    # Trích xuất tiêu đề bài báo
    title = soup.find('h1', class_='title-detail').get_text(strip=True)

    # Trích xuất nội dung bài báo
    # Nội dung thường nằm trong các thẻ p của class 'Normal'
    content_paragraphs = soup.find_all('p', class_='Normal')
    content = "\n".join([para.get_text(strip=True) for para in content_paragraphs])

    # Trích xuất thông tin tác giả
    author_info = soup.find('div', class_='sum-info')
    author = author_info.find('h3', class_='title-news').get_text(strip=True)

    # Hiển thị kết quả
    print("Tiêu đề bài báo:", title)
    print("\nNội dung bài báo:\n", content)
    print("\nTác giả:", author)
else:
    print("Không thể truy cập trang web")
````
